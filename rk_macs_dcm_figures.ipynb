```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set_style('white') #whitegrid
import pandas as pd
import numpy as np
import os
import nibabel as nib
import nilearn
from nilearn import plotting
import glob
from nilearn import masking
```

```python
plotPath = 'D:/MACS/limbiccortical/plots/'

mPFC_coords = [2, 46, -16]
rAmy_coords = [20, -6, -20]
```

## spherical ROIs

```python
from nltools.mask import create_sphere
sphere1 = create_sphere(coordinates=mPFC_coords, radius=12, mask=None)
sphere2 = create_sphere(coordinates=rAmy_coords, radius=8, mask=None)
```

```python
#template_mni = nilearn.datasets.load_mni152_template()
display = plotting.plot_stat_map(sphere2,
                                 colorbar=False,
                                 draw_cross=False,
                                 display_mode = 'z', #'ortho'
                                 cut_coords = [-28,-24,-20,-16,-12,-8],
                                 cmap = plotting.cm.red_transparent,
                                 title="")

display.add_overlay(sphere1,
                    #color = 'blue')
                    cmap=plotting.cm.blue_transparent)
display.savefig(plotPath + "/spheres_z-view.png",dpi=500)
```

## MSDL ROIs

```python
from nilearn import datasets, plotting, image

atlas_data = datasets.fetch_atlas_msdl()
atlas_filename = atlas_data.maps

# First plot the map for the PCC: index 4 in the atlas
display = plotting.plot_stat_map(image.index_img(atlas_filename, 4),
                                 colorbar=False,
                                 draw_cross=False,
                                 title="")

# Now add as an overlay the maps for the ACC and the left and right
# parietal nodes
display.add_overlay(image.index_img(atlas_filename, 5),
                    cmap=plotting.cm.black_blue)
display.add_overlay(image.index_img(atlas_filename, 6),
                    cmap=plotting.cm.black_green)
display.add_overlay(image.index_img(atlas_filename, 3),
                    cmap=plotting.cm.black_pink)

#plotting.show()
display.savefig("D:/RS_DCM_Test_Retest/jupyter/MSDL_plot_ortho.png",dpi=200)
```
